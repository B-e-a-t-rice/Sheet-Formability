```python
import cv2
import numpy as np
```

```python
image = cv2.imread('Sheet 7_000_X001_Y001_G001_A001_I.tiff')
```

```python
cv2.imshow('original',image)
```

```python
kernel3 = np.array([[0, -1,  0],
                   [-1,  5, -1],
                    [0, -1,  0]])
sharp_img = cv2.filter2D(src=image, ddepth=-1, kernel=kernel3)
 
cv2.imshow('Original', image)
cv2.imshow('Sharpened', sharp_img)
     
```
